**sort files in directory base on file names**

```python
# function to sort files in directory base on file names
def sort_file_in_directory(path):
  dir_name =path
  # Get list of all files in a given directory sorted by name
  list_of_files = sorted( filter( lambda x: os.path.isfile(os.path.join(dir_name, x)),
                          os.listdir(dir_name) ) )
  # for file_name in list_of_files:
  #     print(file_name)
  return list_of_files
```
